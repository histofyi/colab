```python
from pymol import cmd, util
from IPython.display import Image

cmd.delete('all')
cmd.load('https://www.histo.fyi/structures/downloads/1hhk_1_abd.cif', 'class_i_alpha')
cmd.load('https://www.histo.fyi/structures/downloads/1hhk_1_peptide.cif', 'peptide')
cmd.reset()
cmd.color('gray80','class_i_alpha')
cmd.hide('cartoon', 'peptide')
cmd.show('sticks', 'peptide')
util.cbay('peptide')
cmd.png('outputs/test.png')
Image(filename='outputs/test.png') 
```
